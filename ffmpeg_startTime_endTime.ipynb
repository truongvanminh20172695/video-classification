```python
import os
import cv2 as cv
import pandas as pd
import numpy as np
import sys

def isNaN(num):
    return num != num

def Time2Frame(strtime,fps):
    hh,mm,ss=strtime.split(':');    
    hh=int(hh)
    mm=int(mm)
    ss=int(ss)
    return fps*(ss+mm*60+hh*3600)

DataDir="A1_460x360"
outputDir="nomalize_A1"
subjectDirs=os.listdir(DataDir)
print(subjectDirs)
fps=30

for subjectDir in subjectDirs:
    # if subjectDir in ['user_id_24026','user_id_24491','user_id_35133','user_id_38058']: #, 'user_id_49381'
    #     continue
    print('subjectDir',subjectDir)
    count=0
    os.makedirs(os.path.join(outputDir,subjectDir), exist_ok=True)
    subject=int(subjectDir.split('_')[-1])
    print('subject',subject)
    subjectPath = os.path.join(DataDir, subjectDir)  
    print('subjectPath',subjectPath)  
    annoFilePath=os.path.join(subjectPath,subjectDir + '.csv')
    print(annoFilePath)
    labeLdata = pd.read_csv(annoFilePath)
    # print(labeLdata)
    # names=('UserID','Filename', 'CameraView','ActivityType','StartTime','EndTime','Label','AppearanceBlock'))
    labeLdata=labeLdata.values
    # print(labeLdata)
    print('labeLdata.shape',labeLdata.shape)   
    print(labeLdata.shape[0]) 
    for i in range (labeLdata.shape[0]):
      
        count +=1
        print('count',count)
        userId=labeLdata[i][0]
        print('userId',userId)
        # if subject !=userId:
        #     sys.exit("Error in csv file %s <> %s" % (subject,userId))            
        fileNeme=labeLdata[i][1].strip()
        # print('fileNeme',fileNeme)
        file_seq=int(fileNeme.split('_')[-1]) 
        # print('file_seq',file_seq)
        videoPath=os.path.join(subjectPath,fileNeme + '.MP4'+'.mp4')
        print('videoPath',videoPath)
        view=labeLdata[i][2].strip()
        print('view',view)
        activityType=labeLdata[i][3]
        if 'Distracted' not in activityType:
            print(activityType)
        startTime=labeLdata[i][4].strip()
        print('startTime',startTime)
        startFrame=Time2Frame(startTime,fps)
        print('startFrame',startFrame)
        endTime=labeLdata[i][5].strip()
        endFrame=Time2Frame(endTime,fps)
        label=labeLdata[i][6]
        try:
            label=int(label)
        except:
            label=18
        print('label',label)        
        appearanceBlock=labeLdata[i][7].strip()
        print('appearanceBlock',appearanceBlock)
        videoOutputName="%03d-%s-%s-%s-%02d-%s.mp4" % (count,file_seq,userId,view,label,appearanceBlock)
        print('videoOutputName',videoOutputName)
        videoOutputFile=os.path.join(outputDir,subjectDir,videoOutputName)
        print('videoOutputFile',videoOutputFile)
        cmd="ffmpeg -n -i %s -ss %s -to %s -vf scale=640:-1 '%s' -hide_banner" % (videoPath,startTime,endTime,videoOutputFile)
        print(cmd)
        os.system(cmd) 
        # print("%s,%d,%s,%d,%s,%d" % (videoOutputFile,userId,view,label,appearanceBlock,int((endFrame-startFrame)/fps)))
```
